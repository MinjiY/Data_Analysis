```python
import requests 
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
url = 'https://news.naver.com/'
req_header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
}
res = requests.get(url, headers= req_header)
print(res.status_code) # 메소드지만 괄호를 주지않음
print(type(res))

print(f'response header = {res.headers}')
print(f'request_header = {res.request.headers}')
```

```python
# res.status_code == 200
if res.ok:
    # Response 객체에서 텍스트 추출
    html = res.text #소스보기 했을때 나오는 코드를 가져오는거 
    #print(html)
    soup = BeautifulSoup(html,'html.parser')
    #print(soup)
    atag_list = soup.select("a[href*='read.nhn']")# read.nhn 이 포함되는 href 의 a태그를 가지고와라
    #print(type(atag_list), len(atag_list))
    # print(atag_list)
    
    for idx, atag in enumerate(atag_list,1):
        title = atag.text.strip()
        link = urljoin(url,atag['href']) # atag속성에 있는 href를 가지고와라
        #link = atag['href']
        
        print(idx, title,link)
        print(title)
```

### Papago 번역 API 호출
- urllib를 사용한 예제 코드 


```python
# 네이버 Papago NMT API 예제
import os                   # os모듈
import sys
import urllib.request
client_id = "" #발급 받았던 id, secret
client_secret = "" 
encText = urllib.parse.quote("Yesterday all my troubles seemed so far away.")
data = "source=en&target=ko&text=" + encText
url = "https://openapi.naver.com/v1/papago/n2mt"
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request, data=data.encode("utf-8"))
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

* requests 사용한 papago 예제코드

```python
import requests

client_id = ""
client_secret = ""
url = "https://openapi.naver.com/v1/papago/n2mt"
enc_text = "Yesterday all my troubles seemed so far away."

req_header = {
    "X-Naver-Client-Id": client_id,
    "X-Naver-Client-Secret": client_secret
}
req_param = {
    "source":"en",
    "target":"ko",
    "text": enc_text
}
res = requests.post(url, data=req_param, headers=req_header)
#print(res.ok, res.status_code)
#print('요청 헤더 ', res.request.headers)
#print('응답 헤더 ', res.headers)

if res.ok:
    print(type(res.text), res.text)
else:
    print('Http status code: ', res.status_code)

# 번역된 문장만 딱 떼내려면? JSON 형태로 받아와야함

```

```python
result_json= res.json()
print(type(result_json), result_json)#dict 타입으로 옴

result_text = result_json['message']['result']['translatedText']  #key를 줘
print(result_text)
```

### 웹툰 Image 다운로드

```python
import requests
import os

req_header = {
    # 웹툰에서 화면에 들어가서 웹툰을 보는 창 의 url
    'referer':'https://comic.naver.com/webtoon/detail.nhn?titleId=703846&no=143&amp;weekday=tue'
}
# 3개 정도만 받음
img_url_list = [
    'https://image-comic.pstatic.net/webtoon/703846/143/20201223165905_bd2183be635280f8706b8caf89a83d53_IMAG01_1.jpg',
    'https://image-comic.pstatic.net/webtoon/703846/143/20201223165905_bd2183be635280f8706b8caf89a83d53_IMAG01_2.jpg',
    'https://image-comic.pstatic.net/webtoon/703846/143/20201223165905_bd2183be635280f8706b8caf89a83d53_IMAG01_3.jpg'
]
# 얘는 text 데이터가 아니라 binary 데이터니까 response에서 content
for img_url in img_url_list:
    res = requests.get(img_url, headers=req_header)
    print(res.ok)
    if res.ok:
        # Response 객체에서 binary 데이터를 가져올때는 content 속성을 사용한다.
        img_data = res.content
        # url 상에서 파일명만 뽑아내서 파일로 저장할꺼야
        file_name = os.path.basename(img_url)
        print(file_name)
        with open(file_name, 'wb') as file:
            print(f'Writing to {file_name} ({len(img_data)} bytes) ')
            file.write(img_data)
        
        
        
        
        
```
